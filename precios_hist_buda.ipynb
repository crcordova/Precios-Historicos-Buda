```python
import time
import datetime
import requests
import pandas as pd
market_id = 'btc-clp'
tiempo = time.time() * 1000
since = datetime.datetime(2021, 10, 1).timestamp() * 1000
df = pd.DataFrame(columns=['date', 'volume','price','tx','id'])
url = f'https://www.buda.com/api/v2/markets/{market_id}/trades'
# loop que va descargando de 100 en 100
while tiempo > since:
    response = requests.get(url, params={
        'timestamp': tiempo,
        'limit': 100,
    })
    df1 = pd.DataFrame(response.json()['trades']['entries'], columns=['date', 'volume','price','tx','id'])
    df = df.append(df1) # Agrego nuevas trades
    tiempo = int(df['date'].min()) # fecha del ultimo dato
    
```
